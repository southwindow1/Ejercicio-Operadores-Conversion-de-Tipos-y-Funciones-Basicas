```python
#Ejercicio 4 Operadores, Conversión de Tipos y Funciones Básicas

data = {}

while True:
    print("Elige una opción:")
    print("1 - Agregar")
    print("2 - Buscar")
    print("3 - Eliminar")
    print("4 - Ver todos los elementos")
    print("5 - Salir")
    print("6 - Calculadora")

    opcion = input("Elige una opción (1 | 2 | 3 | 4 | 5 | 6): ")

    if opcion == '1':
        nombre = input("Escribe el nombre: ")
        telefono = input("Escribe el teléfono: ")
        direccion = input("Escribe la dirección: ")
        data[nombre] = {"telefono": telefono, "direccion": direccion}
        print("Se ha agregado el registro.")

    elif opcion == '2':
        nombre = input("Escribe el nombre: ")
        if nombre in data:
            print(f"El teléfono de {nombre} es {data[nombre]['telefono']} y su dirección es {data[nombre]['direccion']}.")
        else:
            print("No se encontró el nombre en la lista.")

    elif opcion == '3':
        nombre = input("Escribe el nombre: ")
        if nombre in data:
            del data[nombre]
            print("Se ha eliminado el registro.")
        else:
            print("No se encontró el nombre en la lista.")

    elif opcion == '4':
        print("Lista de todos los elementos:")
        for nombre, datos in data.items():
            print(f"{nombre}: {datos['telefono']}, {datos['direccion']}")

    elif opcion == '5':
        print("Gracias por usar el programa.")
        break
    
    elif opcion == '6':
        print("Elige una opción:")
        print("1 - Suma")
        print("2 - Resta")
        print("3 - Multiplicación")
        print("4 - División")
        print("5 - Potencia")
    
        while True:
            operacion = input("Elige una opción (1 | 2 | 3 | 4 | 5): ")
            if operacion in ('1', '2', '3', '4', '5'):
                numero1 = float(input("Escribe un primer número: "))
                numero2 = float(input("Escribe un segundo número: "))
                if operacion == '1':
                    print(f"{numero1} + {numero2} = ", numero1 + numero2)
                elif operacion == '2':
                    print(f"{numero1} - {numero2} = ", numero1 - numero2)
                elif operacion == '3':
                    print(f"{numero1} * {numero2} = ", numero1 * numero2)
                elif operacion == '4':
                    print(f"{numero1} / {numero2} = ", numero1 / numero2)
                elif operacion == '5':
                    print(f"{numero1} ** {numero2} = ", numero1 ** numero2)
    
                break
            else:
                print("Entrada invalida, intenta de nuevo")

    else:
        print("Entrada invalida, intenta de nuevo")
```
